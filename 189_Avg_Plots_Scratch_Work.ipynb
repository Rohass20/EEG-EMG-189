## Averaging emg of different weights across all sessions (all sandpaper)
Sandpaper, 3 different weight classes<br>
Anterior Deltoid 	Brachoradial 	Flexor Digitorum 	Common Extensor Digitorum 	First Dorsal Interosseus

```python
data = funcs.load_participants_ws(1,1,1,9)
```

```python
weight1_emgs = []
weight2_emgs = []
weight3_emgs = []

weightclass1 = 165 #grams
weightclass2 = 330
weightclass3 = 660

#sorting emg data by weight class

for j in range(len(data['participant 1'])):
    
    session = 'session ' + str(j+1)
    
    for i in range(len(data['participant 1'][session])):

        trial = 'trial ' + str(i+1)

        if (data['participant 1'][session][trial]['weight_in_grams'] == weightclass1 and 
            data['participant 1'][session][trial]['texture'] == 'sandpaper' ):

            weight1_emgs.append(data['participant 1'][session][trial]['emg'])

        elif (data['participant 1'][session][trial]['weight_in_grams'] == weightclass2 and 
            data['participant 1'][session][trial]['texture'] == 'sandpaper' ):

            weight2_emgs.append(data['participant 1'][session][trial]['emg'])

        elif (data['participant 1'][session][trial]['weight_in_grams'] == weightclass3 and 
            data['participant 1'][session][trial]['texture'] == 'sandpaper' ):

            weight3_emgs.append(data['participant 1'][session][trial]['emg'])
```

```python
#averaging emg data of each weight class

sum1 = weight1_emgs[0]
sum2 = weight2_emgs[0]
sum3 = weight3_emgs[0]

for i in range(len(weight1_emgs)):
    
    if(i==0): continue
    
    sum1 += weight1_emgs[i]
    
avg1 = sum1/len(weight1_emgs)


for i in range(len(weight2_emgs)):
    
    if(i==0): continue
    
    sum2 += weight2_emgs[i]
    
avg2 = sum2/len(weight2_emgs)

for i in range(len(weight3_emgs)):
    
    if(i==0): continue
    
    sum3 += weight3_emgs[i]
    
avg3 = sum3/len(weight3_emgs)
```

```python
plt.plot(avg1['Anterior Deltoid']),
#plt.plot(avg2['Anterior Deltoid'],alpha = .5),
plt.plot(avg3['Anterior Deltoid'],alpha = .5)
```
